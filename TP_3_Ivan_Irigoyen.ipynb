## Consigna del desafío

1. Vectorizar documentos. Tomar 5 documentos al azar y medir similaridad con el resto de los documentos. Estudiar los 5 documentos más similares de cada uno analizar si tiene sentido la similaridad según el contenido del texto y la etiqueta de clasificación.

2. Transponer la matriz documento-término. De esa manera se obtiene una matriz término-documento que puede ser interpretada como una colección de vectorización de palabras. Estudiar ahora similaridad entre palabras tomando 5 palabras y estudiando sus 5 más similares. La elección de palabras no debe ser al azar para evitar la aparición de términos poco interpretables, elegirlas "manualmente".

3. Entrenar modelos de clasificación Naïve Bayes para maximizar el desempeño de clasificación (f1-score macro) en el conjunto de datos de test. Considerar cambiar parámteros de instanciación del vectorizador y los modelos y probar modelos de Naïve Bayes Multinomial y ComplementNB.

```python
# Import bibliotecas a utilizar
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.metrics import f1_score

# 20newsgroups
from sklearn.datasets import fetch_20newsgroups
import numpy as np
```

## Ejercicio 1: Similitud

```python
# Cargo el dataset completo

newsgroups_all = fetch_20newsgroups(subset='all', remove=('headers','footers','quotes'))
```

```python
# Iniciamos la vectorización

vector = TfidfVectorizer()

# Transformo el vector en una matriz TF-IDF, donde cada fila representa un documento
x_vector = vector.fit_transform(newsgroups_all.data)
```

```python
# Eligo al azar 5 documentos

doc_1, doc_2, doc_3, doc_4, doc_5 = 4547, 54, 464, 7834, 43

docs = [doc_1, doc_2, doc_3, doc_4, doc_5]
```

```python
# Calculo de la similitud de los 5 documentos elegidos.
# El vector lo convertimos en uno de 1D para indexar.
# La similitud la calculamos con la funcion "cosine_similarity()".
# Mostramos los documentos mas similares a cada uno de los elegidos y el valor de la similitud.

for idx, doc in enumerate(docs):
  similitud = cosine_similarity(x_vector[doc], x_vector).flatten() # Vector en 1D
  simili_indices = similitud.argsort()[::-1][1:6] # Los 5 más similares

  print(f"\nDocumento base ({doc}):")
  print("\nTop 5 documentos similares:")
  for simili_idx in simili_indices:
    print(f"  - Documento ({simili_idx}) | Similitud: {similitud[simili_idx]:.4f}")
```

```python
# A que clase pertenecen los documentos originales

print("Clase de los documentos originales:")
for i, doc_index in enumerate(docs):
    print(f"Documento {doc_index}: {newsgroups_all.target_names[newsgroups_all.target[doc_index]]}")
```

```python
# Los 5 documentos con los que comparten mayor similitud.
# Calculo de nuevo la similitud con cosine_similarity(), pero no mostramos su valor,
# con los print llamando a "newsgroups_all.target_names" traemos los documentos a los que pertenece.

for idx, doc in enumerate(docs):
  similitud = cosine_similarity(x_vector[doc], x_vector).flatten() # Vector en 1D
  simili_indices = similitud.argsort()[::-1][1:6] # Los 5 más similares

  print(f"\nDocumento base ({doc}):")
  print("Categoria de los 5 documentos más similares:")
  for simili_idx in simili_indices:
    print(f"  - Documento ({simili_idx}): {newsgroups_all.target_names[newsgroups_all.target[simili_idx]]}")
```

## Ejercicio 2: Transponer matriz y analisis de palabras

```python
# Cargamos los datos y dividimos en train y test

newsgroups_train = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'))
```

```python
# Iniciamos la vectorización

vector = TfidfVectorizer()

# Matriz documento-termino
X_train = vector.fit_transform(newsgroups_all.data)
```

```python
# Transponemos la matriz para que ahora cada columna sea un documento.
# Mostramos la composicion original y la resultante de transponer.

X_t = X_train.T

print(f"Forma original (documento-termino): {X_train.shape}")
print(f"Forma transpuesta (termino-documento): {X_t.shape}")
```

```python
# Conventimos en un array indexable
# Esto devuelve todas las palabras en el orden que estan representadas como columnas de la matriz

terms = np.array(vector.get_feature_names_out())

print(terms)
```

```python
# Seleccionamos las palabras manualmente

palabras = ["space", "windows", "god", "computer", "car"]
```

```python
# Calculo la similitud de las palabras
# Muestra de las palabras elegidas, sus 5 más similares y el valor de la similitud

for word in palabras:
  if word not in terms:
    print(f"'{word}' no esta en el vocabulario del vectorizador")
    continue

  idx = np.where(terms == word)[0][0] # indice de la palabra
  similitud = cosine_similarity(X_t[idx], X_t).flatten() # Calculamos la similitud
  simili_indices = similitud.argsort()[::-1][1:6] # Top 5

  print(f"\nPalabra base: {word}")
  print("Palabras similares")

  for i in simili_indices:
    print(f"{terms[i]} | Similitud: {similitud[i]:.4f}") # Mostramos las similitudes
```

## Ejercicio 3: Naive Bayes

```python
# Definimos y_train donde guardamos los targen que son enteros

y_train = newsgroups_train.target

# Volvemos a definir X_train para que tome los valores del test

X_train = vector.fit_transform(newsgroups_train.data)
```

```python
# Inicio del modelo Naive Bayes Multinominal

clf = MultinomialNB()

clf.fit(X_train, y_train)
```

```python
# Vectorizamos los textos

X_test = vector.transform(newsgroups_test.data)
y_test  = newsgroups_test.target
y_pred = clf.predict(X_test)
```

```python
# Calculamos el F1 score

f1_score(y_test, y_pred, average='macro')
```

```python
# Iniciamos el modelo Naive Bayes ComplementNB

clf2 = ComplementNB()

clf2.fit(X_train, y_train)
```

```python
# Vectorizamos los textos

X_test = vector.transform(newsgroups_test.data)
y_test  = newsgroups_test.target
y_pred = clf2.predict(X_test)
```

```python
# Calculamos el F1 Score

f1_score(y_test, y_pred, average='macro')
```

Conclusión de los modelos de Naive Bayes:

Se ve como el modelo de ComplementNB supera a el MultinominalNB. Ya que nos dio un F1 Score mayor, 0,6929 y 0,5854 respectivamente. Esto nos indica que ComplementNB tiene una mejor capacidad para distinguir entre las diferentes categorias de texto. Sobre todo cuando hay clases con menor representación.
Aunque el F1 Score es alto, nos indica que aún queda margen para la mejora ajustando los distintos parametros o con una mejor vectorización.
